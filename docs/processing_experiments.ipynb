## Configuration

Initial tactic was to stream 10 million rows and get the top 30 most frequent tickers. The first issue was that later on while joining with yfinance, some tickers could not be found for various reasons (name change through the years, simply no data in yfinance). We decided to then look for the top 50 "candidates" and take the first 30 that yfinance could find later.

```python
TOP_N_TICKERS       = 30      # final number of working tickers in the dataset
TICKER_POOL         = 50      # candidates to discover — buffer covers delisted / renamed tickers
ARTICLES_PER_TICKER = 500     # articles to sample per ticker
COUNT_CAP           = 10_000_000  # row limit for the counting pass (reads fast — no storage)
LOAD_CAP            = 3_000_000   # row limit for the loading pass
LABEL_THRESHOLD     = 0.01    # ±1% → up / down / neutral

print(f'Find top {TICKER_POOL} candidates, screen for price data, keep {TOP_N_TICKERS} working tickers')
print(f'Target dataset size: {TOP_N_TICKERS * ARTICLES_PER_TICKER:,} rows')
```

## Section 1 — Load FNSPID

Three-step approach: count tickers across the full dataset, screen for live price data, then load articles only for confirmed working tickers.

- **Pass 1:** streams the full dataset reading only `Stock_symbol` (no storage). Discovers top `TICKER_POOL` (50) candidates by article volume.
- **Price screen:** quick yfinance check on each candidate. Drops delisted / merged / private tickers. Keeps first `TOP_N_TICKERS` (30) that return data.
- **Pass 2:** streams again and samples `ARTICLES_PER_TICKER` rows from each of the 30 confirmed tickers.

Previously the top 30 were loaded first and 7 failed at the price step (DISH, FCAU, GMCR, JWN, MYL, VMW, YHOO). This approach wastes zero articles on inactive tickers.

```python
def count_tickers(cap):
    """Pass 1: count articles per ticker across the full dataset. Streaming instead of downloading. "split=train" has nothing to do with ML, it's just HF labelling.
    Returns (counts dict, total_rows, hit_cap).
    If hit_cap is False, total_rows is the exact dataset size. This was added simply to understand if there were even more rows than what we are scanning.
    But in 10mio rows, we assume that is enough to find information on what the top 50 tickers could be.
    """
    ds = load_dataset('Zihan1004/FNSPID', streaming=True, split='train')
    counts = defaultdict(int)
    total_rows = 0
    for i, row in enumerate(ds):
        total_rows = i + 1
        symbol = row.get('Stock_symbol', '')
        if symbol:
            counts[symbol] += 1
        if i >= cap:
            print(f'Count cap reached at {i:,} rows — dataset has more rows than cap')
            return counts, total_rows, True
        if i % 500_000 == 0 and i > 0:
            print(f'  {i:,} rows scanned — {len(counts):,} unique tickers seen so far')
    print(f'Reached end of dataset at {total_rows:,} rows — this is the exact dataset size')
    return counts, total_rows, False


print('Pass 1: counting articles per ticker...')
all_counts, total_rows, hit_cap = count_tickers(COUNT_CAP)

print(f'\n{"=" * 40}')
if hit_cap:
    print(f'Dataset size: > {COUNT_CAP:,} rows (cap hit — true size unknown)')
else:
    print(f'Dataset size: exactly {total_rows:,} rows')
print(f'Unique tickers found: {len(all_counts):,}')
print(f'{"=" * 40}')

ticker_series     = pd.Series(all_counts).sort_values(ascending=False)
candidate_tickers = list(ticker_series.head(TICKER_POOL).index)

print(f'\nTop {TICKER_POOL} candidate tickers (before price screen):')
print(ticker_series.head(TICKER_POOL).to_string())
```

We can see from the output that by 4.5 mio rows we had already reached the max of unique tickers.

```python
def screen_tickers(candidates, sample_start, top_n):
    """yfinance check: keep the first top_n candidates that return price data."""
    working, skipped = [], []
    for t in candidates:
        if len(working) >= top_n:
            break
        try:
            hist = yf.Ticker(t).history(
                start=sample_start,
                end=sample_start + pd.Timedelta(days=30),
                auto_adjust=False
            )
            if not hist.empty:
                working.append(t)
            else:
                skipped.append(t)
                print(f'  SKIP (no data)   {t}')
        except Exception as e:
            skipped.append(t)
            print(f'  SKIP (error)     {t}: {e}')
    print(f'\nSelected {len(working)} working tickers: {working}')
    if skipped:
        print(f'Skipped {len(skipped)} delisted / renamed / private: {skipped}')
    return working


# Use the start of FNSPID's date range so the screen matches the actual data window
SCREEN_DATE = pd.Timestamp('2012-06-01')

print(f'Screening {len(candidate_tickers)} candidates for yfinance data around {SCREEN_DATE.date()}...')
top_tickers = screen_tickers(candidate_tickers, SCREEN_DATE, TOP_N_TICKERS)
```

```python
fig, ax = plt.subplots(figsize=(14, 4))
bar_colors = ['#5B9BD5' if t in top_tickers else '#B0BEC5' for t in candidate_tickers]
pd.Series(
    [ticker_series[t] for t in candidate_tickers],
    index=candidate_tickers
).plot(kind='bar', ax=ax, color=bar_colors, edgecolor='white')
ax.set_title(
    f'Article Count — Top {TICKER_POOL} Candidates  (blue = selected, grey = skipped/delisted)',
    fontsize=13
)
ax.set_xlabel('Ticker')
ax.set_ylabel('Total Articles in Dataset')
ax.set_xticklabels(candidate_tickers, rotation=45, ha='right', fontsize=8)
plt.tight_layout()
plt.savefig('plots/ticker_coverage.png', dpi=150)
plt.show()
```

We have some important ones that we drop because yfinance has no data on them but as all the other tickers also have a lot of articles available, it is not that big of a deal.

```python
def load_fnspid_top(top_tickers, articles_per_ticker, cap):
    """Pass 2: sample articles from the discovered top tickers."""
    ds = load_dataset('Zihan1004/FNSPID', streaming=True, split='train')
    ticker_counts = defaultdict(int)
    rows = []
    target_set = set(top_tickers)

    for i, row in enumerate(ds):
        symbol = row.get('Stock_symbol', '')
        if symbol in target_set and ticker_counts[symbol] < articles_per_ticker:
            rows.append({
                'date':          row.get('Date', ''),
                'ticker':        symbol,
                'article_title': row.get('Article_title', '')
            })
            ticker_counts[symbol] += 1

        if all(ticker_counts[t] >= articles_per_ticker for t in top_tickers):
            print(f'All tickers reached quota at row {i:,}')
            break

        if i >= cap:
            print(f'Hit load cap ({cap:,} rows). Stopping.')
            break

        if i % 100_000 == 0 and i > 0:
            filled = sum(1 for t in top_tickers if ticker_counts[t] >= articles_per_ticker)
            print(f'  {i:,} rows scanned — {filled}/{len(top_tickers)} tickers complete')

    print(f'\nPer-ticker counts:')
    for t in top_tickers:
        print(f'  {t}: {ticker_counts[t]}')
    return rows


print('Pass 2: loading articles from top tickers...')
raw_rows = load_fnspid_top(top_tickers, ARTICLES_PER_TICKER, LOAD_CAP)
print(f'\nTotal rows loaded: {len(raw_rows):,}')
```

```python
df = pd.DataFrame(raw_rows)

# Clean
df['date'] = pd.to_datetime(df['date'], errors='coerce').dt.strftime('%Y-%m-%d')
df = df.dropna(subset=['date', 'article_title'])
df = df[df['article_title'].str.strip() != '']
df = df.reset_index(drop=True)

# article_id as per data contract: FNSPID_ + zero-padded row number
df.insert(0, 'article_id', ['FNSPID_' + str(i).zfill(5) for i in range(len(df))])

print(f'Shape: {df.shape}')
print(f'Columns: {list(df.columns)}')
print(f'Date range: {df["date"].min()} → {df["date"].max()}')
df.head(3)
```

## Section 3 — Fetch Full Price History from yfinance

> **Skip this section** if `price_cache.pkl` already exists: run only the shortcut cell below, then jump to label assignment.

All `top_tickers` are pre-screened so every fetch succeeds. Full history fetched once per ticker (30 API calls).

```python
if os.path.exists('price_cache.pkl'):
    price_cache = pd.read_pickle('price_cache.pkl')
    missing = [t for t in top_tickers if t not in price_cache]
    if missing:
        print(f'price_cache.pkl exists but is missing {len(missing)} ticker(s): {missing}')
        print('Delete price_cache.pkl and re-run the fetch cell to rebuild for the current ticker list.')
        price_cache = None  # force re-fetch
    else:
        print(f'Loaded price_cache.pkl — all {len(top_tickers)} tickers present')
        print('Skip the fetch cell below. Jump straight to label assignment.')
else:
    price_cache = None
    print('price_cache.pkl not found. Run the fetch cell below.')
```

```python
def fetch_price_cache(tickers, date_min, date_max):
    """Fetch full price history once per ticker. Returns {ticker: Series of Close prices}."""
    cache = {}
    failed = []
    for ticker in tickers:
        try:
            hist = yf.Ticker(ticker).history(start=date_min, end=date_max)
            if hist.empty:
                print(f'  EMPTY  {ticker}')
                failed.append(ticker)
                continue
            # Handle both tz-aware and tz-naive DatetimeIndex returned by yfinance
            if hist.index.tz is not None:
                hist.index = hist.index.tz_convert(None).normalize()
            else:
                hist.index = hist.index.normalize()
            cache[ticker] = hist['Close']
        except Exception as e:
            print(f'  FAILED {ticker}: {e}')
            failed.append(ticker)
    print(f'\nCached {len(cache)} / {len(tickers)} tickers. Failed: {failed or "none"}')
    return cache


date_min = pd.to_datetime(df['date'].min()) - pd.Timedelta(days=5)
date_max = pd.to_datetime(df['date'].max()) + pd.Timedelta(days=10)

print(f'Fetching prices for {len(top_tickers)} tickers ({date_min.date()} → {date_max.date()})...')
price_cache = fetch_price_cache(top_tickers, date_min, date_max)
```

### Label Assignment

Uses the price cache to compute T and T+1 closing prices for each article, calculates the percentage change, and assigns a directional label (up / down / neutral) using the `LABEL_THRESHOLD` defined in Configuration.

```python
def get_t_and_t1(ticker, date_str, cache):
    """Return (price_t, price_t1) or (None, None) if data unavailable."""
    prices = cache.get(ticker)
    if prices is None or prices.empty:
        return None, None
    date = pd.Timestamp(date_str)
    future = prices.index[prices.index >= date]
    if len(future) < 2:
        return None, None
    return float(prices[future[0]]), float(prices[future[1]])


def assign_label(pct_change, threshold):
    if pct_change > threshold * 100:
        return 'up'
    elif pct_change < -threshold * 100:
        return 'down'
    return 'neutral'


print('Assigning prices and labels...')
results = [get_t_and_t1(row['ticker'], row['date'], price_cache) for _, row in df.iterrows()]

df['price_t']    = [round(r[0], 2) if r[0] else None for r in results]
df['price_t1']   = [round(r[1], 2) if r[1] else None for r in results]
df['pct_change'] = [
    round((r[1] - r[0]) / r[0] * 100, 4) if r[0] and r[1] else None
    for r in results
]
df['label'] = [
    assign_label(p, LABEL_THRESHOLD) if pd.notna(p) else None
    for p in df['pct_change']
]

before = len(df)
df = df.dropna(subset=['label']).reset_index(drop=True)
print(f'Dropped {before - len(df):,} rows (no matching price data)')
print(f'Final dataset: {len(df):,} rows')
df[['article_id', 'date', 'ticker', 'article_title', 'price_t', 'price_t1', 'pct_change', 'label']].head(3)
```

### Outlier Analysis — Relabeling Without Extreme Price Moves

Some rows have very large price swings (e.g. earnings crashes, buyout spikes) that are not driven by the kind of news sentiment FinBERT can learn from. 
Removing them before training may give the classifier a cleaner signal.

We use the **IQR 1.5× rule**: any row where `pct_change` falls outside `[Q1 − 1.5×IQR, Q3 + 1.5×IQR]` is flagged as an outlier. 
The graphs below compare the label distribution before and after removal.

```python
# IQR-based outlier detection
Q1 = df['pct_change'].quantile(0.25)
Q3 = df['pct_change'].quantile(0.75)
IQR = Q3 - Q1
lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR

df_no_outliers = df[(df['pct_change'] >= lower_fence) & (df['pct_change'] <= upper_fence)].copy()
n_removed = len(df) - len(df_no_outliers)

print(f'IQR fence: [{lower_fence:.2f}%, {upper_fence:.2f}%]')
print(f'Rows removed: {n_removed:,} ({n_removed / len(df) * 100:.1f}%)')
print(f'Rows remaining: {len(df_no_outliers):,}')

label_counts_orig  = df['label'].value_counts().reindex(['up', 'down', 'neutral'])
label_counts_clean = df_no_outliers['label'].value_counts().reindex(['up', 'down', 'neutral'])
colors = ['#009E73', '#D55E00', '#B0BEC5']

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Original — bar
label_counts_orig.plot(kind='bar', ax=axes[0, 0], color=colors, edgecolor='white')
axes[0, 0].set_title(f'With Outliers  (n={len(df):,})', fontsize=13)
axes[0, 0].set_xlabel('')
axes[0, 0].set_ylabel('Count')
axes[0, 0].set_xticklabels(['up', 'down', 'neutral'], rotation=0)
for i, v in enumerate(label_counts_orig):
    axes[0, 0].text(i, v + 30, f'{v / len(df) * 100:.1f}%', ha='center', fontsize=10)

# Original — pie
label_counts_orig.plot(kind='pie', ax=axes[0, 1], colors=colors, autopct='%1.1f%%', startangle=90)
axes[0, 1].set_title('Label Share', fontsize=13)
axes[0, 1].set_ylabel('')

# No outliers — bar
label_counts_clean.plot(kind='bar', ax=axes[1, 0], color=colors, edgecolor='white')
axes[1, 0].set_title(f'Without Outliers  (n={len(df_no_outliers):,},  fence [{lower_fence:.1f}%, {upper_fence:.1f}%])', fontsize=13)
axes[1, 0].set_xlabel('')
axes[1, 0].set_ylabel('Count')
axes[1, 0].set_xticklabels(['up', 'down', 'neutral'], rotation=0)
for i, v in enumerate(label_counts_clean):
    axes[1, 0].text(i, v + 30, f'{v / len(df_no_outliers) * 100:.1f}%', ha='center', fontsize=10)

# No outliers — pie
label_counts_clean.plot(kind='pie', ax=axes[1, 1], colors=colors, autopct='%1.1f%%', startangle=90)
axes[1, 1].set_title('Label Share', fontsize=13)
axes[1, 1].set_ylabel('')

plt.suptitle('Label Distribution: With vs. Without Outliers (IQR 1.5×)', fontsize=14, y=1.02)
plt.tight_layout()
plt.savefig('plots/label_distribution_outlier_comparison.png', dpi=150, bbox_inches='tight')
plt.show()

print('\n=== Majority-class baseline comparison ===')
print(f'With outliers:    {label_counts_orig.max() / len(df) * 100:.1f}%  (majority class: neutral)')
print(f'Without outliers: {label_counts_clean.max() / len(df_no_outliers) * 100:.1f}%  (majority class: neutral)')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

sns.set_theme(style='whitegrid')

cal_df = pd.read_csv('processed_data.csv')
print(f'Loaded processed_data.csv: {len(cal_df):,} rows, {cal_df["ticker"].nunique()} tickers')
print(f'pct_change range: {cal_df["pct_change"].min():.2f}% → {cal_df["pct_change"].max():.2f}%')
cal_df[['pct_change', 'label']].describe()
```

### Experiments 1 & 2

```python
CANDIDATE_THRESHOLDS = [0.005, 0.01, 0.015, 0.02]  # 0.5 %, 1 %, 1.5 %, 2 %

def relabel(pct, t):
    if pct > t * 100:  return 'up'
    if pct < -t * 100: return 'down'
    return 'neutral'

# Experiment 1: class distribution at each threshold 
records = []
for t in CANDIDATE_THRESHOLDS:
    labels = cal_df['pct_change'].apply(lambda p: relabel(p, t))
    counts = labels.value_counts(normalize=True) * 100
    records.append({
        'threshold': f'±{t*100:.1f}%',
        'up':      counts.get('up',      0),
        'down':    counts.get('down',    0),
        'neutral': counts.get('neutral', 0),
        'majority_baseline': counts.max()
    })
    print(f'  ±{t*100:.1f}%  →  up {counts.get("up",0):.1f}%  down {counts.get("down",0):.1f}%  neutral {counts.get("neutral",0):.1f}%  '
          f'(majority baseline {counts.max():.1f}%)')

sweep_df = pd.DataFrame(records).set_index('threshold')

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

# stacked bar
colors = {'up': '#009E73', 'down': '#D55E00', 'neutral': '#B0BEC5'}
bottom = np.zeros(len(sweep_df))
for label, color in colors.items():
    axes[0].bar(sweep_df.index, sweep_df[label], bottom=bottom, label=label,
                color=color, edgecolor='white', width=0.5)
    bottom += sweep_df[label].values
axes[0].set_title('Class Distribution by Threshold', fontsize=13)
axes[0].set_ylabel('% of dataset')
axes[0].set_xlabel('Threshold')
axes[0].legend(loc='upper right', fontsize=9)

# majority baseline line
axes[1].plot(sweep_df.index, sweep_df['majority_baseline'], marker='o', color='#5B9BD5', linewidth=2)
axes[1].set_title('Majority-Class Baseline vs Threshold', fontsize=13)
axes[1].set_ylabel('Accuracy if always predicting majority class (%)')
axes[1].set_xlabel('Threshold')
axes[1].yaxis.set_major_formatter(mticker.FormatStrFormatter('%.1f%%'))

plt.suptitle('Experiment 1 — Threshold Sweep', fontsize=14, y=1.02)
plt.tight_layout()
plt.savefig('plots/threshold_sweep.png', dpi=150, bbox_inches='tight')
plt.show()

# Experiment 2: return distribution with threshold lines overlaid 
fig, ax = plt.subplots(figsize=(12, 4))
clipped = cal_df['pct_change'].clip(-15, 15)
ax.hist(clipped, bins=100, color='#5B9BD5', edgecolor='white', alpha=0.85)

line_styles = ['-', '--', '-.', ':']
for (t, ls) in zip(CANDIDATE_THRESHOLDS, line_styles):
    ax.axvline( t * 100, color='#009E73', linestyle=ls, linewidth=1.5, label=f'+{t*100:.1f}%')
    ax.axvline(-t * 100, color='#D55E00', linestyle=ls, linewidth=1.5, label=f'-{t*100:.1f}%')
ax.axvline(0, color='black', linewidth=0.8, alpha=0.4)
ax.set_title('Experiment 2 — Return Distribution with Candidate Thresholds (clipped ±15%)', fontsize=13)
ax.set_xlabel('% Change T → T+1')
ax.set_ylabel('Count')
ax.legend(ncol=4, fontsize=8)
plt.tight_layout()
plt.savefig('plots/threshold_distribution.png', dpi=150)
plt.show()
```

### Experiment 3 — Threshold Sweep Without Outliers (IQR 1.5×)

Same as Experiments 1 & 2 above, but run on the outlier-removed dataset (IQR fence: [−5.12%, +5.13%], 1,458 rows removed, 13,527 remaining).

This shows whether removing extreme price moves changes which threshold is optimal and what the new majority-class baseline looks like.

```python
# Apply IQR filter to the calibration dataset
_Q1 = cal_df['pct_change'].quantile(0.25)
_Q3 = cal_df['pct_change'].quantile(0.75)
_IQR = _Q3 - _Q1
_lower = _Q1 - 1.5 * _IQR
_upper = _Q3 + 1.5 * _IQR

cal_clean = cal_df[(
    cal_df['pct_change'] >= _lower) & (cal_df['pct_change'] <= _upper
)].copy()

print(f'IQR fence: [{_lower:.2f}%, {_upper:.2f}%]')
print(f'Rows removed: {len(cal_df) - len(cal_clean):,}  |  Rows remaining: {len(cal_clean):,}')
print()

# Experiment 3: same threshold sweep on cleaned data
records_clean = []
print('Threshold sweep on outlier-removed data:')
for t in CANDIDATE_THRESHOLDS:
    labels = cal_clean['pct_change'].apply(lambda p: relabel(p, t))
    counts = labels.value_counts(normalize=True) * 100
    records_clean.append({
        'threshold': f'±{t*100:.1f}%',
        'up':      counts.get('up',      0),
        'down':    counts.get('down',    0),
        'neutral': counts.get('neutral', 0),
        'majority_baseline': counts.max()
    })
    print(f'  ±{t*100:.1f}%  →  up {counts.get("up",0):.1f}%  down {counts.get("down",0):.1f}%  '
          f'neutral {counts.get("neutral",0):.1f}%  (majority baseline {counts.max():.1f}%)')

sweep_clean = pd.DataFrame(records_clean).set_index('threshold')

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

# Stacked bar
colors_map = {'up': '#009E73', 'down': '#D55E00', 'neutral': '#B0BEC5'}
bottom = np.zeros(len(sweep_clean))
for label, color in colors_map.items():
    axes[0].bar(sweep_clean.index, sweep_clean[label], bottom=bottom,
                label=label, color=color, edgecolor='white', width=0.5)
    bottom += sweep_clean[label].values
axes[0].set_title('Class Distribution by Threshold (no outliers)', fontsize=13)
axes[0].set_ylabel('% of dataset')
axes[0].set_xlabel('Threshold')
axes[0].legend(loc='upper right', fontsize=9)

# Majority baseline — no outliers vs. original side by side
sweep_orig = pd.DataFrame(records).set_index('threshold')  # from Experiment 1
x = np.arange(len(sweep_clean))
w = 0.35
axes[1].bar(x - w/2, sweep_orig['majority_baseline'],  width=w, label='With outliers',    color='#5B9BD5', edgecolor='white')
axes[1].bar(x + w/2, sweep_clean['majority_baseline'], width=w, label='Without outliers', color='#F0A500', edgecolor='white')
axes[1].set_title('Majority-Class Baseline: With vs. Without Outliers', fontsize=13)
axes[1].set_ylabel('Accuracy if always predicting majority class (%)')
axes[1].set_xlabel('Threshold')
axes[1].set_xticks(x)
axes[1].set_xticklabels(sweep_clean.index)
axes[1].yaxis.set_major_formatter(mticker.FormatStrFormatter('%.1f%%'))
axes[1].legend(fontsize=9)

plt.suptitle('Experiment 3 — Threshold Sweep Without Outliers', fontsize=14, y=1.02)
plt.tight_layout()
plt.savefig('plots/threshold_sweep_no_outliers.png', dpi=150, bbox_inches='tight')
plt.show()

print('\n=== Summary: majority baseline shift from outlier removal ===')
for t in sweep_orig.index:
    orig = sweep_orig.loc[t, 'majority_baseline']
    clean = sweep_clean.loc[t, 'majority_baseline']
    print(f'  {t}:  {orig:.1f}%  →  {clean:.1f}%  (Δ {clean-orig:+.1f}pp)')
```
